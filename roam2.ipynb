```python
#pgroups
%matplotlib inline
plt.rcParams["figure.figsize"] = [10,7]
def plotProducts(pds):
    data = []
    for p in pds:
        mysales = train2[(train2["product_key"] == p)]
        mysales = mysales[mysales["store_id"] == 137]
        mysales = mysales.sort_values(by="calendar_date")
        #mysales.index.get_level_values("calendar_date").values
        data.append(mysales)
    #print(len(data)) 
    for d in data:
        dates = d["calendar_date"].values
        #print(dates)
        x = []
        for dt in dates:
            dts = dt.split("/")
            thisDate = datetime(int(dts[2]), int(dts[0]), int(dts[1]), 0,0,0)
            #print(thisDate)
            x.append(thisDate)
        y = d["total_sales"].values
        #print("hi", x)
        plt.plot(x,y)
    #plt.rcParams["figure.figsize"] = [40,40]
    plt.show()
    
plotProducts([14, 41])
```
